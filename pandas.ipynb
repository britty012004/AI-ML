```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'john', 'mary'],
    'Age': [25, 30, 38, 40]
}

df = pd.DataFrame(data)

df.drop('Age', axis=1, inplace=True)

print(df)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'john', 'mary']
})

df2 = pd.DataFrame({
    'key': [1, 2, 3, 4],
    'Age': [25, 30, 38, 40]
})

result = pd.merge(df1, df2, on='key')
print(result)
```
